```python
%matplotlib inline

import csv
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
```

##Generation vs Irradiance scatterplot

```python
generation = []
irradiance = []

f = open("data_solaire_minute_over_day.csv", "r")

headers = open("Headers2.txt", "w")
for row in csv.reader(f):
    for i in range(0, len(row)):
        headers.write(row[i] + '\n')
    break
headers.close()

time = 0
for row in csv.reader(f):
    irradiance.insert(time, float(row[24]))
    generation.insert(time, float(row[2]))
    time=time+1
    
plt.scatter(irradiance,generation)
plt.ylabel("Generation [kW]")
plt.xlabel("Irradiance [W/m^2]")
plt.title('Generation (kW) and Irradiance (W/m^2) in a day')
plt.show()
#plt.savefig('Generation and Irradiance in a day.png')
```

##Regression (run after scatterplot is plotted)

```python
val =(len(generation)==len(irradiance))
#print("Good length? "+ str(val))

X_training = np.array(irradiance)
X_training = np.reshape(X_training,(len(X_training),1))
#print("shape of X_training")
#print(np.shape(X_training))

Y_training = np.array(generation)
Y_training = np.reshape(Y_training,(len(Y_training),1))
#print("shape of Y_training")
#print(np.shape(Y_training))

regr = lm.LinearRegression() #classifier
regr.fit(X_training,Y_training)


#USING SAME TRAINING SET AS TEST SET

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(X_training) - Y_training) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X_training, Y_training))

# Plot outputs
plt.scatter(X_training, Y_training,  color='black')
plt.plot(X_training, regr.predict(X_training), color='blue',
         linewidth=2)
plt.ylabel("Generation [kW]")
plt.xlabel("Irradiance [W/m^2]")
plt.title("Predictor of Generation wrt Irradiance (Linear Regression)")

plt.show()
#plt.savefig('Generation and Irradiance in a day with Linear Regression.png')
```

####Import data of day over month as test set

```python
generation2 = []
irradiance2 = []

f = open("data_solaire_minute_over_month.csv", "r")

for row in csv.reader(f):
    break
    
time = 0
for row in csv.reader(f):
    irradiance2.insert(time, float(row[24]))
    generation2.insert(time, float(row[2]))
    time=time+1

    
X_test = np.array(irradiance2)
X_test = np.reshape(X_test,(len(X_test),1))

Y_test = np.array(generation2)
Y_test = np.reshape(Y_test,(len(Y_test),1))

#USING SET OVER MONTH AS TEST SET (same time each day)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(X_test) - Y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X_test, Y_test))

# Plot outputs
plt.scatter(X_test, Y_test,  color='black')
plt.plot(X_test, regr.predict(X_test), color='blue',
         linewidth=2)
plt.ylabel("Generation [kW]")
plt.xlabel("Irradiance [W/m^2]")
plt.title("LR Predictor of Generation wrt Irradiance (over month)")

#plt.show()
plt.savefig('Generation and Irradiance over month with Trained Linear Regression.png')
```
